```python
from transformers import AutoTokenizer
```

```python
tok = AutoTokenizer.from_pretrained("./tokenizers/morph-wp")
```

```python
cases = [
    "свежевскопанный",
    "свежевскопанная",
    "свежевскопанные",
    "свежевскопанными",
    "свежевскопанных",

    "свежеспрятанный",
    "свежеспрятанная",
    "свежеспрятанные",
    "свежеспрятанными",
    "свежеспрятанных",

    "!Linguistics!"
]
```

```python
for case in cases:
    print(f"{case} -> {tok.tokenize(case)}")
```

```python
tok2 = AutoTokenizer.from_pretrained("./tokenizers/morph-sigm-wp")
```

```python
for case in cases:
    print(f"{case} -> {tok2.tokenize(case)}")
```

```python
tok3 = AutoTokenizer.from_pretrained("./tokenizers/morph-wp2")
```

```python
for case in cases:
    print(f"{case} -> {tok3.tokenize(case)}")
```
